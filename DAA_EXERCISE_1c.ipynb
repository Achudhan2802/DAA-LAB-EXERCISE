```python
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= key <= arr[high]:

        comparisons += 1

        if arr[high] == arr[low]:
            if arr[low] == key:
                return low, comparisons
            return -1, comparisons

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < low or pos > high:
            break

        comparisons += 1

        if abs(arr[pos] - key) < 1e-9:
            return pos, comparisons

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons
import random

sizes = [10000, 50000, 100000]

for size in sizes:
    arr = sorted(random.uniform(0.0, 1000.0) for _ in range(size))

    key = arr[size // 2]      # Search for middle element

    index, comparisons = interpolation_search(arr, key)

    print(f"Size = {size}")
    print(f"Index Found = {index}")
    print(f"Comparisons = {comparisons}")
    print()
```
